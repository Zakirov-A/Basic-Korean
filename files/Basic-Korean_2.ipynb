У согласных букв есть `имя`

Согласные были созданы путем подражания речи

`*` - мягкость

`ㄱ` = к* (к*иёк)

`가` = для вертикальных

`고` = для горизонтальных

`구` = 9 (ку)

`고기` - мясо) (коги)

`ㄴ` = н (ниын)

`냐` = для вертикальных

`뇨` = для горизонтальных

`누나` = сестра для мальчиков (нуна)

`ㄷ` = т* (т*игыт)

`더` = для вертикальных

`두` = для горизонтальных

`도로` = дорога (торо)

`구두` = туфли (куту)

`ㄹ` = р (риыль)

`거리` = улица (кори)

`다리` = ноги (тари)

`ㅁ` = м (миым)

`어머니` = мама (амани)

`머리` = голова или волосы (мори)

`ㅂ` = п* (п*иыб)

`바다` = море (пада)

`나비` = бабочка (наби)

`ㅅ` = c* (cиот)

>с комбанициями глассных `a` `о` `оу` `у` `ы` читается `с`

>с комбинациями глассных `йа` `йо` `йоу` `йу` `йи` читается `щ`

`소나무` = сосна (сонаму)

`ㅈ` = ть (тьиыт)

`자다` = cпать (тьада)
