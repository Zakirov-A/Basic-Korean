## Двойные Согласные

`ㄲ` = мягкая К (санкиёк)

`ㄸ` = мягкая Т (сантигыт)

`ㅃ` = мягкая П (санпиып)

`ㅆ` = мягкая С (санщиёт)

`ㅉ` = буква т (сантиыт)

### ПОДСЛОГИ

> `Подчим` = согласная буква, которая находится под слогами

`물` = вода (муль)

`달` = дочка (таль)

`교실` = аудитория (кёошиль)

`볼링` = боулинг (поллинг)

> В словах где кончается на согласные (кх, к, мягкая к) читаем как `к`

`수박` арбуз (субак)

`부앜` кухня (буак)

`밖` снаружи (бак)

`눈` глаза, снег (нун)

`친구` друг (чингу)

`그림` картина (кырин)

`사람` человек (сарам)

`이름` имя (ирым)

`집` дом (тьип)

`앞` стул (ап)

`빵` хлеб (панг)

`방` комната (паанг)

`사랑` любовь (саранг)

> В словах которые заканчиваются на согласные (т, тх, с,мягкая(удвоенная) с, ть, ч) читаем как  `Т`

`곧` = скоро (кот)

`옷` = одежда (от)

`낮` = день (нат)

`꽃` = цветок (кот)
